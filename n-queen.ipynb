```python
Queens = [-1]*8
Queens
```

```python
def isSafe(queens, row, col):
    for row_others, col_others in zip(range(len(queens)),queens):
        if col_others == -1:
            break
        if col_others == col:
            return False
        diffx = abs(row - (row_others+1))
        diffy = abs(col - col_others)
        if diffx == diffy:
            return False
    return True
```

```python
def solveNQueen(queens, row):

    print(queens)
    if row == 9: 
        return True
    for col in range(1, 9):
        if isSafe(queens, row, col):
            queens[row-1] = col
            if solveNQueen(queens.copy(), row+1):
                return True
    return False
```

```python
solveNQueen(Queens, 1)
```
